# 2021-06-09 BubbleSet 的な可視化を実現できそうな仕掛けを試した

```python
import numpy as np

from bokeh.models import LogColorMapper
from bokeh.plotting import figure, output_notebook, show

output_notebook()

def normal2d(X, Y, sig=[0.1, 0.1], mu=[0.0, 0.0]):
    z = ((X - mu[0])**2 + (Y - mu[1])**2) / (sig[0]**2 + sig[1]**2)
    return np.exp(-z/2) / np.sqrt(2 * np.pi * sig[0] * sig[1])

X, Y = np.mgrid[-3:3:100j, -2:2:100j]
Z = normal2d(X, Y, mu=[1.0, 1.0]) + normal2d(X, Y, mu=[1.0, -0.2]) + normal2d(X, Y, mu=[-0.2, 1.0])
image = Z * 1e6

color_mapper = LogColorMapper(palette="Greys256", low=1, high=1e7)

plot = figure(x_range=(0, 1), y_range=(0, 1), toolbar_location=None)
plot.image(image=[image], color_mapper=color_mapper,
           dh=[1.0], dw=[1.0], x=[0], y=[0])

show(plot)
```
